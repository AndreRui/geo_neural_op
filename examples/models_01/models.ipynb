```python
import torch
import numpy as np
import os
from pathlib import Path
import pyvista as pv

if Path.cwd().as_posix().endswith("/models_01"):
    os.chdir("../..")
from gnp.models import GNP, PatchGNP
```

```python
# Instantiating GNP with the full kernel GraphConvolution
full_model = GNP(
    node_dim=3,
    edge_dim=6,
    out_dim=1,
    layers=[64] * 10,
    conv_name="GraphConvolution",
    conv_args={"neurons": 128},
    nonlinearity="ReLU",
    skip_connection=True,
    device="cuda",
)
```

```python
# Instantiating GNP with the block-factorized kernel using 4 blocks/channels
block_model = GNP(
    node_dim=3,
    edge_dim=6,
    out_dim=1,
    layers=10 * [64],
    conv_name="BlockFactorizedConvolution",
    conv_args={"neurons": 128, "num_channels": 4},
    nonlinearity="ReLU",
    skip_connection=True,
    device="cuda",
)
```

```python
# Instantiating GNP with the separable full kernel
separable_full_model = GNP(
    node_dim=3,
    edge_dim=3,  # must be same as node_dim
    out_dim=1,
    layers=10 * [64],
    conv_name="SeparableConvolution",
    conv_args={"neurons": 128, "kernel_name": "FullKernel", "kernel_args": {}},
    nonlinearity="ReLU",
    skip_connection=True,
    device="cuda",
)
```

```python
# Instantiating GNP with the separable full kernel
separable_block_model = GNP(
    node_dim=3,
    edge_dim=3,  # must be same as node_dim
    out_dim=1,
    layers=10 * [64],
    conv_name="SeparableConvolution",
    conv_args={
        "neurons": 128,
        "kernel_name": "BlockKernel",
        "kernel_args": {"num_channels": 4},
    },
    nonlinearity="ReLU",
    skip_connection=True,
    device="cuda",
)
```

```python
#Instantiating PatchGNP model
patch_model = PatchGNP(
    node_dim=3,
    out_dim=16,
    layers=5*[64],
    num_channels=4,
    neurons=256,
    nonlinearity="ReLU",
    device="cuda"
)
```
